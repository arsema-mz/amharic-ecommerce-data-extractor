```python
import pandas as pd

# Load the combined messages dataset
df = pd.read_csv('../data/combined_messages.csv')

print(df.head())
```

```python
# Select a random subset of 30 messages
subset_messages = df['message'].sample(n=30, random_state=1).tolist()

# Alternatively, select the first 30 messages
# subset_messages = df['message'].head(30).tolist()

# Display the selected messages
for i, message in enumerate(subset_messages):
    print(f"{i+1}: {message}\n")
```

# Create the CoNLL Format for Labeling
Now, you need to prepare the selected messages for labeling in the CoNLL format:

## Define the CoNLL Format Structure:
Each token in a message will be labeled according to its entity type. You can initialize an empty list to store the CoNLL-formatted data.

## Tokenization:
Split each message into individual tokens (words).

## Labeling:
For now, weРђЎll set the labels to "O" (outside) as placeholders, which you can later update when you manually label the entities.

```python
# Initialize an empty list to store CoNLL formatted data
conll_data = []

# Define a function to convert messages to CoNLL format
def to_conll_format(message):
    tokens = message.split()  # Simple tokenization by whitespace
    for token in tokens:
        conll_data.append(f"{token}\tO")  # Placeholder label "O"
    conll_data.append("")  # Blank line to separate sentences/messages

# Convert each message in the subset to CoNLL format
for message in subset_messages:
    to_conll_format(message)

# Display a portion of the CoNLL formatted data
print("\n".join(conll_data[:50]))  # Display the first 50 lines
```
